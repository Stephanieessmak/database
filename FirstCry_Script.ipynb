```python
scopes = ['https://www.googleapis.com/auth/spreadsheets', 'https://www.googleapis.com/auth/drive']
json = '/content/drive/MyDrive/Essmak Data Work/Requirements/Credentials.json'
credentials = Credentials.from_service_account_file(json, scopes=scopes)

gc = gspread.authorize(credentials)

gauth = GoogleAuth()
drive = GoogleDrive(gauth)

df = gc.open_by_url("https://docs.google.com/spreadsheets/d/1jdCfmvWAuRnyPVfR7lx38-jXKNeSOno-vmKnU1fjhk4/edit#gid=497917819")
sku = df.worksheet('Combined SKU Details')
sku=pd.DataFrame(sku.get_all_records()) 

# reading cost sheet data
cp = gc.open_by_url('https://docs.google.com/spreadsheets/d/18wCjME5w01pB3TTIXcQaq1KnrtiTORrqkUHiNIzycZk/edit#gid=227041125')
cp = cp.worksheet('Cost Sheet')
cp =pd.DataFrame(cp.get_all_records())

ff = gc.open_by_url('https://docs.google.com/spreadsheets/d/18wCjME5w01pB3TTIXcQaq1KnrtiTORrqkUHiNIzycZk/edit#gid=227041125')
ff = ff.worksheet('Franchise Fee') 
ff =pd.DataFrame(ff.get_all_records())
ff=ff.loc[ff['Platform']=='FirstCry']

count = 0

# open a google sheet
gs = gc.open_by_url('https://docs.google.com/spreadsheets/d/1I8NdsQyAWgI7p237Hj7OsBou56IA1wYUHrXvuOFRj6M/edit#gid=0')
# select a work sheet from its name
worksheet1 = gs.worksheet('FirstCry')


# data = pd.read_csv('/content/drive/MyDrive/Essmak Data Work/CRM /order-data_2023-03-11.csv')
df = gc.open_by_url("https://docs.google.com/spreadsheets/d/1durdZwaOeXVA1AeYYwUTeg2Y6BPgyaWJhzWhl8RAn4E/edit#gid=0")
AED = df.worksheet('AED To USD')
AED=pd.DataFrame(AED.get_all_records()) 




# spreadsheet3 =  pd.read_excel( "/content/drive/MyDrive/Essmak Data Work/Abbreviation/Abbrevations.xlsx",sheet_name='Sheet1')
# AED = pd.read_excel("/content/drive/MyDrive/Essmak Data Work/Currency Values/Currency.xlsx", sheet_name = 'AED to USD')

AED['Date']=AED['Date'].astype(str)
AED['Date']=AED['Date'].apply(lambda x: x.split(' ')[0])

AED['Date'] = pd.to_datetime(AED['Date'])
AED['Date'] = AED['Date'].dt.strftime('%d-%m-%Y')


cp['Date'] = pd.to_datetime(cp['Date'])
cp['Product'] = cp['Product'].apply(lambda x: x.title().lower())

for i in os.listdir('/content/drive/MyDrive/Essmak Data Work/FC Invoice'):
  main_data = pd.read_csv('/content/drive/MyDrive/Essmak Data Work/Database/merged_data_file.csv')
  pdf_file_names = pd.read_csv('/content/drive/MyDrive/Essmak Data Work/Database/pdf_names_file.csv')
  
  
  a = pd.DataFrame()
  all_files_in_drive = list()
  
  if i.endswith('.pdf') and i not in pdf_file_names['Filename'].to_list():
    file_path = '/content/drive/MyDrive/Essmak Data Work/FC Invoice/' + str(i)
    dfs = tabula.read_pdf(file_path, pages='all', relative_area=True, relative_columns=True, multiple_tables=False,
                     area = [0,0,550,100], columns = [6,20,30,43,55,69,73,85], pandas_options={'header':None})
    df2 = dfs[0][13:-6]
    new_header = df2.iloc[0] #grab the first row for the header
    df2 = df2[1:]
    df2.columns = new_header
    file_name_repeater = list()
    file_name_repeater.extend(repeat(str(i),len(df2['SrNo'])))
    df2['invoice_pdf_name'] = file_name_repeater
    main_data = main_data.append(df2)
    main_data.to_csv('/content/drive/MyDrive/Essmak Data Work/Database/merged_data_file.csv',index=False)
    # # dataframe (create or import it)


    all_files_in_drive.append(i)
    a['Filename'] = all_files_in_drive
    pdf_file_names = pdf_file_names.append(a)
    pdf_file_names.to_csv('/content/drive/MyDrive/Essmak Data Work/Database/pdf_names_file.csv',index=False)
    # main_

    # new = main_data["Style Code"].str.split("-", n=4, expand = True)
    # main_data['Category_code'] = new[0]   
    # print(main_data) 
    result = pd.merge(main_data, sku, how='left', left_on = "Style Code" , right_on = "Product SKU") 
    
  



    # df_values = df2.values.tolist()
    # result = result.rename(columns={'Full form': 'Product Category'})
    # result['Product Category'] = result.apply(lambda row: row['Product Category'] if pd.notna(row['Product Category']) else row['Category_code'], axis=1) 
    

    # result['Product Category']= result['Product Category'].replace(np.nan, "No SKU Details")
    result['Date'] = pd.to_datetime(result['Order Date'])


    result['day'] = result['Date'].dt.day
    result['month'] = result['Date'].dt.month
    result['year'] = result['Date'].dt.year



    result['Order Date'] = pd.to_datetime(result['Order Date'], format='%Y/%m/%d').dt.strftime('%d-%m-%Y')
    # result['O']



    result = result.merge(AED,
                  left_on= 'Order Date',
                  right_on = 'Date',
                  how= 'left')
    # result.head()
    result['Total (Excl. Tax)']=result['Total (Excl. Tax)'].astype(float)
    result['Close']=result['Close'].astype(float)
    result['Total USD'] = result['Total (Excl. Tax)']*   result['Close']
    # result=result.drop(['Category_code', 'Abbrevation', 'Close','Date'], axis=1)
    # result["Order Date"] =  pd.to_datetime(result['Order Date'],format='%Y/%m/%d')
    # result['Order Date'] = result['Order Date'].dt.strftime('%d-%m-%Y')
    result['Qty'] = result['Qty'].astype('int')

    
    # result[['day', 'month', 'year']] = result['Order Date'].str.split('-', expand=True)
    result['Country'] = 'United Arab Emirates'

    result= pd.merge(result, cp, how='left', left_on = ['Product_Category','day', 'month', 'year'] , right_on = ["Product", 'day', 'month', 'year'])
    
    result= pd.merge(result, ff, how='left', left_on = "year" , right_on = "Year") 
    result['Franchise fees'] = result['Franchise fees'].str.replace('%', '').astype(float)
    result['Franchise fees_2'] = result['Franchise fees'] / 100

    result['Franchise_amount'] = result['Total USD'] * result['Franchise fees_2']

    
    df = pd.DataFrame(result)
    df['Total Cost'] = np.where(df['Cost Price'].isnull(), np.nan, df['Cost Price'] + df['Franchise_amount'])


    # Cost_Price=result['Cost Price'].to_list()
    # Franchise_amount=result['Franchise_amount'].to_list()
    # total_amount=[]
    # for i in range(len(Cost_Price)):
    #   if Cost_Price[i]==0:
    #     total_amount.append(Cost_Price[i])
    #   else:
    #     total_amount.append(Cost_Price[i]+Franchise_amount[i])
    # result['Total cost']=total_amount
    
    df['Profit'] = np.where(df['Cost Price'].isnull(), np.nan, df['Total USD'] - df['Total Cost'])
    df=df.drop('Platform', axis=1)

    df=df.drop(['Date_x','Date_y'] ,axis=1)

    df = df.rename(columns={'Franchise fees':'Franchise fees_percent', 'Franchise fees_2':'Franchise fees_2_percent'})

    worksheet1.clear()
    set_with_dataframe(worksheet=worksheet1, dataframe=df, include_index=False,include_column_header=True, resize=True)
    count =  count + 1
    print(count)

```
